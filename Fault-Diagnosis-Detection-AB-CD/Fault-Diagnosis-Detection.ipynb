```python
X_full = data.iloc[:, [3,4,6,7,8]].values
```

```python
X_full
```

```python
y_full = data['type_of_fault'].values
```

```python
y_full
```

```python
sc = StandardScaler()
```

```python
sc.fit(X_full)
```

```python
X_full = sc.transform(X_full)
```

```python
X_full
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_full, y_full, test_size = 0.25, random_state = 1)
```

```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
classifier = Sequential()
classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu', input_dim = 5))
classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
yhat_train = (classifier.predict(X_train) > 0.5)
yhat_test = (classifier.predict(X_test) > 0.5)
```

```python
pred_dnn = classifier.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python
def build_dnn_classifier():
    classifier = Sequential()
    classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu', input_dim = 3))
    classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier

classifier = KerasClassifier(build_fn = build_dnn_classifier, batch_size = 10, epochs = 100)

scores = cross_val_score(estimator = classifier,X = X_train, y = y_train, cv = 10, n_jobs = 1)
```

```python
accuracy = metrics.accuracy_score(y_train, yhat_train)
confusion_matrix = metrics.confusion_matrix(y_train, yhat_train)
classification = metrics.classification_report(y_train, yhat_train)
print()
print('============================== DNN Model Evaluation ==============================' )
print()
print ("Cross Validation Mean Score:" "\n", scores.mean())
print()
print ("Model Accuracy:" "\n", accuracy)
print()
print("Confusion matrix:" "\n", confusion_matrix)
print()
print("Classification report:" "\n", classification) 
print()
```
